# Storytelling

This has been my first data analysis project in the Iron Hack Bootcamp.

Our task was to make an analysis based on the data of a csv file of shark attacks around the world. We were asked to think about a couple of hypotheses and to study them based on the data. 

In order to do so, I used the libraries `seaborn`, `pandas`, `matplotlib`, `numpy` and `re`. 

As this is my first time doing an analysis using python pandas, it was a challenging experience and I faced Many difficulties. 

On the following lines, I will explain the results of my research, and how I reached my conclusions. 

## 1. Are sex and age factors that influence the possibility of an incident with a shark?

During the preliminary analysis of the data, the information that most catch my attention was the categorization between the 'Provoked' and 'Unprovoked' attacks. 

My mind inmediately thought about the possible relations between the sex and the exposure to risks facing sharks. 

Therefore, I decided to base my first hypotheses on it. 

### 1st Hypothesis: Men have more incidents with sharks

The first graph shows that, according to the data, men have both more provoked and unprovoked incidents with sharks. In this case, a provoked incident would be an encounter initiated by a person, and the unprovoked an unexpected attack of the shark. 

### 2nd Hypothesis: Boys and young men have more incidents with sharks

In my opinion this was the most intesting part of the analysis. The figure above shows that the young men between 15 and 21 years old have the most incidents. 

The big difference between men and women across all ages is also noteworthy. 

### 3rd Hypothesis: Young men have more fatal incidents with sharks

Following the trend of the previous charts, this one also shows that men have more fatal accidents than women.
