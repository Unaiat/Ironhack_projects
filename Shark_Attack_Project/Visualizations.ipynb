```python
df_p_u = pd.read_csv("Data/df_p_u.csv",encoding = "ISO-8859-1") 
```

```python
# Graph showing the amount of provoked and unprovoked attacks by sex
fig, ax = plt.subplots(figsize = (8, 8))
p_u_attacks = sns.countplot(x = df_p_u['P/U'], hue = df_p_u['Sex'], palette="Blues")
ax.set_title("Unprovoked and provoked attacks (by sex)")
plt.show()
```

```python
vi.save_graph(p_u_attacks, "images/p_u_attacks.svg")
```

```python
fig, ax = plt.subplots(figsize = (12, 8))
a_of_vic = sns.countplot(x = 'Activity of victim', data = df_calculation2, palette="Set2")
ax.set_title("Activity realized by a victim of an attack (non-fatal attacks)")
plt.show()
```

```python
vi.save_graph(a_of_vic, "images/a_of_vic.svg")
```

```python
# This shows the amout of people attacked that survived, by activity and sex
fig, ax = plt.subplots(figsize = (12, 8))
a_of_vic_s = sns.countplot(x = 'Activity of victim', hue = 'Sex', data = df_calculation2, palette="Set2")
ax.set_title("Activity realized by a victim of an attack (by sex / non-fatal attacks)")
plt.show()
```

```python
vi.save_graph(a_of_vic_s, "images/a_of_vic_s.svg")
```
